# Exploratory Data Analysis of Health Factors Influencing Cognitive Decline

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("cognitive_decline_dataset.csv")
df.head()
```

```python
print(df.info())
print(df.describe())
```

```python
sns.countplot(x="CognitiveDecline", data=df)
plt.title("Cognitive Decline Distribution")
plt.show()
```

```python
sns.histplot(df["Age"], kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
sns.boxplot(x="CognitiveDecline", y="MemoryScore", data=df)
plt.title("Memory Score vs Cognitive Decline")
plt.show()
```

```python
sns.boxplot(x="CognitiveDecline", y="PhysicalActivity", data=df)
plt.title("Physical Activity vs Cognitive Decline")
plt.show()
```

```python
sns.boxplot(x="CognitiveDecline", y="SleepHours", data=df)
plt.title("Sleep Hours vs Cognitive Decline")
plt.show()
```

```python
sns.boxplot(x="CognitiveDecline", y="BMI", data=df)
plt.title("BMI vs Cognitive Decline")
plt.show()
```

```python
sns.heatmap(df.corr(), annot=True)
plt.title("Feature Correlation")
plt.show()
```

### Key Insights

- Individuals with lower memory scores are more likely to experience cognitive decline  
- Reduced physical activity is associated with higher cognitive risk  
- Sleep duration shows moderate influence on cognitive health  
- Age has a positive correlation with cognitive decline  

### Conclusion

This analysis highlights the importance of lifestyle and cognitive factors in understanding and predicting cognitive health.
