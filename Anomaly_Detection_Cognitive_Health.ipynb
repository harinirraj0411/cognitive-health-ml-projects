# Anomaly Detection in Health Data for Identifying High-Risk Cognitive Cases

This project uses anomaly detection techniques to identify unusual patterns in health data that may indicate high-risk cognitive conditions.


```python
import pandas as pd
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("cognitive_decline_dataset.csv")
df.head()
```

```python
model = IsolationForest(contamination=0.1, random_state=42)

df["Anomaly"] = model.fit_predict(df.drop("CognitiveDecline", axis=1))
```

```python
df["Anomaly"].value_counts()
```

```python
plt.scatter(df["Age"], df["MemoryScore"], c=df["Anomaly"])
plt.title("Anomaly Detection (Age vs Memory Score)")
plt.xlabel("Age")
plt.ylabel("Memory Score")
plt.show()
```

### Key Insights

- Anomaly detection identified a subset of individuals with unusual health patterns  
- These anomalies may represent high-risk cognitive cases  
- Machine learning can help in early identification of abnormal health conditions  

### Conclusion

Anomaly detection techniques provide a useful approach for identifying high-risk individuals and can complement predictive models in healthcare analysis.
